## Data preprocessing

As RecipeBox constitutes only a small part of entire dataset, we are allowing it to have measurements of the ingredients - to have some 
new style (noise) in the dataset. 

1) Remove punctuations from the ingredients in RecipeNLG "->None \\
2) Removal of new line character from the RecipeBox \\ 
3) Remove recipes with no ingredients or instructions \\
4) Remove all the recipes that have are too short and too long instructions \\
5) Remove all the redundant recipes (Code present, but session crashes) \\


```python
def preprocess_NLG(x):
  #Removing all the inverted commas 
  return x.replace('"','')
```

```python
#Removing the square brackets and quotes from instructions of RecipeNLG
def preprocess_NLG_instructions(x):
  x = x[2:-2]
  x = x.replace('", "',' ')
  return x
```

```python
#Relacing all the next line character tokens with a space
def preprocess_RB_instructions(x):
  return x.replace('\n',' ')
```

```python
#Function to check whether an entry is empty or None
def check_nan(x):
  if x==[] or x==None or x=='':
    return True
  else:
    return False
```

```python
#Removing all the recipes which have either no instructions listed, or no ingredients mentioned
def remove_incomplete_recipes(list_of_dfs):
  for i in range(len(list_of_dfs)):
    t1 = list_of_dfs[i]['instructions'].apply(check_nan).values
    t2 = list_of_dfs[i]['ingredients'].apply(check_nan).values
    t = t1 | t2
    labels = list(compress(range(len(t)), t))
    list_of_dfs[i].drop(index=labels, inplace=True)
    list_of_dfs[i].reset_index(drop=True, inplace=True)
  return list_of_dfs
```

```python
#Calculates the length of instruction based on number of words in it
def len_instructions(x):
  try:
    return len(x.split())
  except:
    return 0
```

```python
#Removes all the recipes that have either very long instructions or very short length instructions
#Short instrcutions are removed based on the threshold - can modify it to increase or decrease the size of dataset
def remove_short_long_recipes(list_of_dfs, count, threshold):
  for i in range(len(list_of_dfs)):
    t = list_of_dfs[i]['instructions'].apply(len_instructions).values
    t_excess_len= t>count[i]
    t_less_len = t<threshold
    labels = list(compress(range(len(t_excess_len)), t_excess_len)) + list(compress(range(len(t_less_len)), t_less_len))
    list_of_dfs[i].drop(labels, inplace=True)
    list_of_dfs[i].reset_index(drop=True,inplace=True)
  return list_of_dfs
```

```python
#Calculates average length on an instruction in each database
def avg_recipe_length(list_of_dfs, threshold):
  count=[]
  for i in range(len(list_of_dfs)):
    lengths = list_of_dfs[i]['instructions'].apply(len_instructions)
    count.append(sum(lengths.values)/len(list_of_dfs[i]))
  return remove_short_long_recipes(list_of_dfs,count, threshold)
```

```python
#Removes all the redundant recipes from the dataset if they have similarity score of more than the threshold
def remove_redundant(list_of_dfs, cosine_similarity, threshold):
  labels = []
  mask = (cosine_similarity > threshold & cosine_similarity<1).any(axis=1)
  labels = cosine_similarity[mask]
  final_df = pd.concat([list_of_dfs[0],list_of_dfs[1]], ignore_index=True)
  final_df.drop(labels, inplace=True)
  final_df.reset_index(drop=True, inplace=True)
  return final_df
```

```python
#Calculating the similarity between recipes of both databases
#Cosine similarity is calculated between the vectorized versions obtained by applying TF-IDF word vectorization scheme
import numpy as np
def calculate_similarity(list_of_dfs, threshold):
  recipe_list = list_of_dfs[0].instructions.values.tolist() + list_of_dfs[1].instructions.values.tolist()
  vect = TfidfVectorizer(min_df=1, stop_words="english")                                                                                                                                                                                                   
  
  tfidf = vect.fit_transform(recipe_list)                                                                                                                                                                                            
  cos_similarity = cosine_similarity(tfidf,tfidf)   
  return remove_redundant(list_of_dfs, cos_similarity, threshold)
```

```python
#Performing data preprocessing
list_of_dfs[0]['ingredients'] = list_of_dfs[0]['ingredients'].apply(preprocess_NLG)
list_of_dfs = remove_incomplete_recipes(list_of_dfs)
list_of_dfs = avg_recipe_length(list_of_dfs,50)
list_of_dfs[0].instructions = list_of_dfs[0].instructions.apply(preprocess_NLG_instructions)
list_of_dfs[1].instructions = list_of_dfs[1].instructions.values.tolist()
list_of_dfs[1].instructions = list_of_dfs[1].instructions.apply(preprocess_RB_instructions)
# final_df = calculate_similarity(list_of_dfs,0.7)
```
