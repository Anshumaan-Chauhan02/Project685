##Extarcting the Recipe Generation Dataset

```python
#Importing pandas to load the datasets
import pandas as pd
```

```python
#Unzipping RecipeBox and RecipeNLG datasets
!unzip '/content/drive/MyDrive/685 Project/dataset.zip'
!unzip '/content/drive/MyDrive/685 Project/recipes_raw.zip'
```

```python
#Loading the datasets into pandas Dataframe
rnlg_df = pd.read_csv('/content/dataset/full_dataset.csv')
rb_df1 = pd.read_json('/content/recipes_raw_nosource_ar.json')
rb_df2 = pd.read_json('/content/recipes_raw_nosource_epi.json')
rb_df3 = pd.read_json('/content/recipes_raw_nosource_fn.json')
```

```python
#Preprocessing the datasets to get them into a common format
rb_df1 = rb_df1.T
rb_df2 = rb_df2.T
rb_df3 = rb_df3.T

#Changing the names of the columns
rnlg_df = rnlg_df[['title','NER','directions']]
rnlg_df.rename(columns={'NER':'ingredients','directions':'instructions'}, inplace=True)

#Selecting only the necessary columns
rb_df1 = rb_df1[['ingredients','instructions']]
rb_df2 = rb_df2[['ingredients','instructions']]
rb_df3 = rb_df3[['ingredients','instructions']]

#Combining the RecipeBox dataset into one single dataframe
rb_df = pd.concat([rb_df1,rb_df2,rb_df3])

#Deleting earlier dataframes to save memory
del rb_df1
del rb_df2
del rb_df3

#Reset the value of index for rb_df
rb_df.reset_index(drop=True, inplace=True)
```

##Statistics of the dataset before preprocessing 

```python
#Defining a function to get number of words in the input
def avg_length(x):
  return  len(str(x).split())
```

```python
#Storing average length of the input in separate columns
rb_df['len'] = rb_df['ingredients'].apply(avg_length) + rb_df['instructions'].apply(avg_length)
rnlg_df['len'] = rnlg_df['ingredients'].apply(avg_length) + rnlg_df['instructions'].apply(avg_length)
```

```python
#Number of examples before preprocessing in each dataset
print(len(rb_df))
print(len(rnlg_df))
```

```python
#Max-Min-Avg length of the input in RecipeNLG
rnlg_df['len'].describe()
```

```python
#Max-Min-Avg length of the input in RecipeBox
rb_df['len'].describe()
```
